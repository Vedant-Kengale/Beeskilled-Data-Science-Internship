```python
# Import the libraries
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("StudentsPerformance.csv")

# Check first 5 rows
print(df.head())
```

```python
# display dataset information
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
# Clean Missing data

# 1] Remove missing data
df = df.dropna()

# # 2] Fill Missing Numeric value
# df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Calculate Average Scores

# 1] Math Score Average
print('Average Maths Score: ',df['math score'].mean())

# 2] Reading Score Average
print('Average Reading Score: ',df['reading score'].mean())

# 3] Writing Score Average
print('Average Writing Score: ',df['writing score'].mean())
```

```python
# Calculate Maximum Scores

# 1] Highest Math Score
print('Highest Maths Score: ',df['math score'].max())

# 2] Highest Reading Score
print('Highest Reading Score: ',df['reading score'].max())

# 3] Highest Writing Score
print('Highest Writing Score: ',df['writing score'].max())
```

```python
# Calculate Minimum Scores

# 1] Lowest Math Score
print('Lowest Maths Score: ',df['math score'].min())

# 2] Lowest Reading Score
print('Lowest Reading Score: ',df['reading score'].min())

# 3] Lowest Writing Score
print('Lowest Writing Score: ',df['writing score'].min())
```

```python
# Display Results
print(df.describe())
```
